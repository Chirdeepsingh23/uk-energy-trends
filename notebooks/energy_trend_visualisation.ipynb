# UK Energy Mix Analysis (Hourly, National Grid)

This notebook explores how the United Kingdom's electricity generation mix changes over time using hourly data from the **Carbon Intensity API**.

We will:
- Clean and prepare the raw generation mix data
- Analyze how different fuel sources behave across time (hour, day, season)
- Visualise renewable vs fossil trends
- Identify dominant fuel types and their contributions

By the end, we‚Äôll draw data-driven insights on how the UK grid is powered throughout the year ‚Äî from solar peaks to winter gas demand.

> üìä Tools: Python, Pandas, Matplotlib  
> üìÅ Source: `generation_mix_history.csv` (hourly UK electricity generation mix)


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load long-format historical fuel data
df = pd.read_csv("../data/generation_mix_history.csv")

df.head(9)
```

## Data Cleaning & Preprocessing

Before we analyze the fuel data, we apply several cleaning steps to ensure the dataset is reliable:

- üßπ Drop rows with missing timestamp, fuel type, or percentage share
- üìÖ Convert timestamps to datetime format
- üßæ Remove exact duplicate rows
- ‚è±Ô∏è Round timestamps to the nearest hour (since some values may be slightly offset)
- üßÆ Group by hour and fuel type, averaging the percentage in case of duplicates from overlapping API calls

This ensures each row represents a **unique fuel type at a specific hour**, with clean and consistent percentage values.


```python
# Remove nulls in required columns (precaution)
df = df.dropna(subset=["timestamp", "fuel", "perc"])

# Convert timestamp column to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Drop duplicate rows (exact matches)
df = df.drop_duplicates()

# Round timestamps to nearest hour (floored)
df["timestamp_hour"] = df["timestamp"].dt.floor("h")

# Group by timestamp + fuel to resolve multiple entries (e.g. from API overlaps)
# Take mean of 'perc' if duplicates exist
clean_df = df.groupby(["timestamp_hour", "fuel"], as_index=False)["perc"].mean()

clean_df.head(9)
```

## Data Quality Check

Now that the data is cleaned and grouped, we perform a quick data quality check to verify that:

- All hourly timestamps are present
- There are no remaining null values
- The total percentage share across all fuels per hour is reasonable (i.e., close to 100%)

These checks help confirm that the dataset is ready for accurate time-series analysis.


```python
# Count missing timestamps
print("Unique hourly timestamps:", clean_df["timestamp_hour"].nunique())

# Check if any NaNs left
print("Any remaining NaNs:", clean_df.isna().sum().sum())

# Check total fuel per hour
totals = clean_df.groupby("timestamp_hour")["perc"].sum()
print("Min fuel total per hour:", totals.min())
print("Max fuel total per hour:", totals.max())
```

```python
# Pivot to wide format: one row per hour
pivot_df = clean_df.pivot(index="timestamp_hour", columns="fuel", values="perc")

# Fill missing fuel values with 0 (not present = not used)
pivot_df = pivot_df.fillna(0)

# Sort just in case
pivot_df = pivot_df.sort_index()
```

## Normalise Fuel Shares and Extract Hour of Day

We now normalise each row of the dataset so that the percentage values across all fuel types **sum to 100%** per hour.  
This ensures fair comparisons and correct visualisation scaling.

We also extract the **hour of the day** (0‚Äì23) from the timestamp so we can analyse daily patterns in fuel usage later on.


```python
# Normalize each row to total 100%
pivot_df_normalized = pivot_df.div(pivot_df.sum(axis=1), axis=0) * 100

# Rounding
pivot_df_normalized = pivot_df_normalized.round(2)


# Add a column for hour of day (0‚Äì23)
pivot_df_normalized["hour"] = pivot_df_normalized.index.hour
```

## Hour-of-Day Fuel Share Analysis

We calculate the average percentage share of each fuel type for every hour of the day (0‚Äì23).  
This reveals daily patterns in energy generation ‚Äî for example, when solar peaks or when gas fills in demand.

To do this, we:
- Exclude metadata columns (like `hour`, `weekday`, etc.)
- Group the normalised fuel data by hour
- Compute the mean share for each fuel type across all days


```python
# Get only fuel columns (exclude 'hour')
fuel_cols = pivot_df.columns.intersection(pivot_df_normalized.columns)
numeric_fuels = pivot_df_normalized[fuel_cols].copy()

# Add back hour column to fuel data
numeric_fuels["hour"] = pivot_df_normalized["hour"]

# Group by hour and calculate mean share
hourly_avg = numeric_fuels.groupby("hour").mean()
```

## Insight 1: Hourly Patterns in UK Fuel Share

The chart below illustrates the average share of each fuel type across every hour of the day.

Notable observations:
-  **Solar** rises from 6 AM and peaks around midday, then drops sharply in the evening
-  **Gas** compensates during early mornings and evenings when solar dips, showing classic demand-following behaviour
-  **Wind** is fairly consistent throughout the day, with a slight increase overnight
-  **Nuclear** remains mostly steady, contributing as a reliable baseload
-  **Imports** and **biomass** remain relatively flat, with minor shifts across the day

This highlights how the UK grid dynamically balances generation from renewable and non-renewable sources based on time-of-day demand and availability.


```python
# Plot the average fuel share by hour
hourly_avg.plot(figsize=(14, 6))
plt.title("Average Fuel Share by Hour of Day (UK)")
plt.xlabel("Hour of Day (0 = Midnight, 12 = Noon)")
plt.ylabel("Average Share (%)")
plt.grid(True)
plt.legend(title="Fuel Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Stacked Fuel Mix by Hour

To visualise how all fuel types contribute collectively across a 24-hour period, we plot a stacked area chart.  
This chart shows the combined makeup of the UK energy mix by hour, with each colour representing a fuel type's average share.
It also visually reinforces how gas and solar complement each other throughout the day, while wind and nuclear maintain steady contributions.


```python
# Stacked area chart of average fuel share by hour
hourly_avg.plot(kind="area", stacked=True, figsize=(14, 6), alpha=0.8)

plt.title("Stacked Fuel Share by Hour of Day (UK)")
plt.xlabel("Hour of Day (0 = Midnight, 12 = Noon)")
plt.ylabel("Total Fuel Share (%)")
plt.grid(True)
plt.legend(title="Fuel Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

plt.show()
```

## Day of the week Analysis

```python
# Add weekday name (e.g., Monday, Tuesday) to each row
pivot_df_normalized["weekday"] = pivot_df_normalized.index.day_name()
```

## Calculate Average Fuel Share by Weekday

Now that each row is labelled with a weekday, we group the data by day of the week to calculate the average share of each fuel type.

Steps:
- Remove non-fuel columns such as `hour` and `total`
- Retain the weekday label for grouping
- Calculate the mean percentage share of each fuel type for each day

This will help us understand whether energy generation patterns differ between weekdays, such as during working days versus weekends.


```python
# Only fuel columns (exclude 'hour', 'weekday' etc.)

columns_to_exclude = ["hour", "total", "weekday"]
fuel_cols = [col for col in pivot_df_normalized.columns if col not in columns_to_exclude] # Drop non-fuel columns explicitly
fuel_data = pivot_df_normalized[fuel_cols].copy()

# Add weekday back in for grouping
fuel_data["weekday"] = pivot_df_normalized["weekday"]

# Group by weekday and take mean
weekday_avg = fuel_data.groupby("weekday").mean()
```

```python
# Reorder to start from Monday
ordered_days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_avg = weekday_avg.reindex(ordered_days)
```

## Insight 2: Average Fuel Share by Day of Week

This chart shows how the UK's energy mix varies across different days of the week.

Key observations:
- Gas usage tends to be slightly higher during the midweek (especially Wednesday), likely due to greater industrial activity.
- Wind appears more dominant towards the weekend, especially on Sundays.
- Solar and nuclear remain relatively stable, unaffected by the working calendar.
- Imports rise slightly on weekends, potentially reflecting adjustments in domestic generation demand.

This weekday analysis highlights how grid usage and generation shift subtly between working and non-working days.


```python
# Drop non-fuel columns again before plotting
columns_to_exclude = ["hour", "total", "weekday", "day_type", "month", "season"]
plot_data = weekday_avg.drop(columns=columns_to_exclude, errors="ignore")

# Plot
plot_data.plot(figsize=(14, 6))
plt.title("Average Fuel Share by Day of Week (UK)")
plt.ylabel("Average Share (%)")
plt.xlabel("Weekday")
plt.grid(True)
plt.legend(title="Fuel Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

## Weekday vs Weekend Fuel Share

To compare broader behavioural patterns in energy generation, we classify each timestamp as either:

- **Weekday** (Monday to Friday)
- **Weekend** (Saturday and Sunday)

We then group the data by this `day_type` and calculate the average share of each fuel type. This allows us to see how generation strategies differ based on demand cycles between the working week and the weekend.

```python
# Add a 'day_type' column: Weekday or Weekend
pivot_df_normalized["day_type"] = pivot_df_normalized.index.to_series().dt.dayofweek.apply(
    lambda x: "Weekend" if x >= 5 else "Weekday"
)
```

```python
# Keep only fuel columns
columns_to_exclude = ["hour", "total", "weekday", "day_type"]
fuel_cols = [col for col in pivot_df_normalized.columns if col not in columns_to_exclude]

# Prepare fuel data
fuel_data = pivot_df_normalized[fuel_cols].copy()
fuel_data["day_type"] = pivot_df_normalized["day_type"]

# Group and calculate mean for weekdays and weekends
daytype_avg = fuel_data.groupby("day_type").mean()
```

## Insight 3: Fuel Mix Comparison - Weekday vs Weekend

This bar chart compares the average fuel share between weekdays and weekends for each energy source.

Key takeaways:
- **Gas usage** is significantly higher during weekdays, likely reflecting commercial and industrial demand.
- **Wind** plays a larger role on weekends, possibly aided by lower demand and more favourable weather conditions.
- **Imports** and **nuclear** both see a slight increase on weekends, potentially to compensate for reduced domestic fossil generation.
- Other sources such as **biomass**, **solar**, and **hydro** show relatively minor variation.

This comparison highlights how energy demand patterns affect fuel selection, with weekends leaning more towards cleaner and consistent sources.


```python
# Transpose for bar plot (so each fuel gets its own group)
daytype_avg.T.plot(kind="bar", figsize=(12, 6))
plt.title("Average Fuel Share: Weekday vs Weekend")
plt.ylabel("Average Share (%)")
plt.xlabel("Fuel Type")
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title="Day Type")
plt.tight_layout()
plt.show()
```

## Monthly Fuel Share Aggregation

To understand how energy generation changes over the course of the year, we extract the month name from each timestamp and calculate the average fuel share for each month.

Steps:
- Label each row with its respective calendar month
- Remove metadata columns to isolate fuel types
- Group the data by month and calculate the mean share
- Reorder months chronologically to maintain logical readability

This analysis will help us detect broader seasonal trends, such as increased solar generation in summer or higher gas usage in colder months.


```python
# Add month name (e.g., January, February)
pivot_df_normalized["month"] = pivot_df_normalized.index.month_name()
```

```python
# Exclude non-fuel columns
columns_to_exclude = ["hour", "total", "weekday", "day_type", "month"]
fuel_cols = [col for col in pivot_df_normalized.columns if col not in columns_to_exclude]

# Extract fuel data and add month
fuel_data = pivot_df_normalized[fuel_cols].copy()
fuel_data["month"] = pivot_df_normalized["month"]

# Group by month and take mean
monthly_avg = fuel_data.groupby("month").mean()
```

```python
# Ensure months are in calendar order
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
monthly_avg = monthly_avg.reindex(month_order)
```

## Insight 4: Average Fuel Share by Month

This line chart presents how the share of each fuel type varies across calendar months.

Insights:
- **Solar** output increases from March to July, peaking during summer months before tapering off in autumn.
- **Wind** generation tends to be higher during winter months, with dips around spring and summer.
- **Gas** fills in the demand during colder months, especially between November and March.
- **Imports** and **nuclear** remain relatively steady, although imports show minor variation around spring and summer.

These monthly trends reflect both environmental factors (e.g., sunlight, temperature, wind conditions) and seasonal electricity demand.


```python
# Plot trends for each fuel type across months
monthly_avg.plot(figsize=(14, 6))
plt.title("Average Fuel Share by Month (UK)")
plt.xlabel("Month")
plt.ylabel("Average Share (%)")
plt.grid(True)
plt.legend(title="Fuel Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Seasonal Fuel Mix Aggregation

To explore long-term patterns in electricity generation, we categorise each data point into a meteorological season:

- **Winter**: December, January, February  
- **Spring**: March, April, May  
- **Summer**: June, July, August  
- **Autumn**: September, October, November

Steps:
- Map each timestamp to its corresponding season
- Group the fuel data by season
- Transpose the result so that fuel types become the index (rows) and seasons the columns

This allows us to create clear comparisons across seasons for each fuel source, aiding in the visual interpretation of cyclical generation patterns.


```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

# Apply the season mapping to each timestamp
pivot_df_normalized["season"] = pivot_df_normalized.index.month.map(get_season)
```

```python
# Filter just fuel columns again
columns_to_exclude = ["hour", "total", "weekday", "day_type", "month", "season"]
fuel_cols = [col for col in pivot_df_normalized.columns if col not in columns_to_exclude]

fuel_data = pivot_df_normalized[fuel_cols].copy()
fuel_data["season"] = pivot_df_normalized["season"]

# Group and average
seasonal_avg = fuel_data.groupby("season").mean()
```

```python
# seasonal_avg: seasons as rows, fuels as columns
# We'll transpose it for plotting
seasonal_avg = seasonal_avg.reindex(["Winter", "Spring", "Summer", "Autumn"])
seasonal_avg_T = seasonal_avg.T  # fuels as rows, seasons as columns
```

## Insight 5: Seasonal Comparison of Fuel Share

This grouped bar chart shows the average share of each fuel type across the four seasons.

Observations:
- **Gas** usage is highest in winter and autumn, reflecting its role in meeting heating demand.
- **Wind** peaks strongly in winter and remains significant throughout the year, highlighting its seasonal strength.
- **Solar** shows a sharp peak in summer and spring, as expected.
- **Imports** and **nuclear** vary moderately but appear to increase slightly in summer and spring.
- Other fuels like **biomass** and **hydro** maintain consistent but minor roles across all seasons.

This seasonal view adds important context to the monthly trends and confirms expected patterns related to weather and energy demand cycles.


```python
# Plot: x = fuel types, grouped bars = seasons
ax = seasonal_avg.T.plot(
    kind="bar", 
    figsize=(12, 6), 
    color=["#1f77b4", "#2ca02c", "#ffcc00", "#d17a22"],  # Blue, Green, Yellow, Autumn Brown
)

plt.title("Average Fuel Share by Season (UK)")
plt.xlabel("Fuel Type")
plt.ylabel("Average Share (%)")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.legend(["Winter", "Spring", "Summer", "Autumn"], title="Season")
plt.tight_layout()
plt.show()

```

## Clean vs Fossil Fuel Mix (Monthly Trend)

To assess the UK's progress towards sustainable energy, we group fuel types into:

- **Clean**: wind, solar, hydro, nuclear  
- **Fossil**: gas, coal

We calculate the combined monthly average share for each group. This high-level comparison allows us to visualise the shift (if any) between clean and fossil-based generation over time, supporting long-term decarbonisation analysis.


```python
# Define categories
clean_fuels = ["wind", "solar", "hydro", "nuclear"]
fossil_fuels = ["gas", "coal"]

# Create a new DataFrame with clean/fossil groupings
fuel_mix_groups = pd.DataFrame(index=pivot_df_normalized.index)

# Sum grouped values
fuel_mix_groups["clean"] = pivot_df_normalized[clean_fuels].sum(axis=1)
fuel_mix_groups["fossil"] = pivot_df_normalized[fossil_fuels].sum(axis=1)
```

```python
# Add month column
fuel_mix_groups["month"] = fuel_mix_groups.index.to_series().dt.to_period("M")

# Group by month and take mean
monthly_mix = fuel_mix_groups.groupby("month")[["clean", "fossil"]].mean()
```

## Insight 6: Monthly Trend: Clean vs Fossil Fuel Share

This line chart compares the monthly average share of clean versus fossil fuel sources in the UK's energy generation mix.

Key insights:
- **Clean energy consistently dominates**, with its share remaining above 45% in most months.
- Peaks in clean generation align with summer (solar) and winter (wind) surges.
- **Fossil fuel usage** rises during colder months (notably November‚ÄìJanuary), likely due to heating demand.
- A slight inverse trend is visible ‚Äî when fossil share rises, clean share dips, and vice versa ‚Äî indicating how one group compensates for the other.

This view provides a high-level assessment of energy strategy balance over time and underscores the seasonal pressures influencing clean vs fossil reliance.


```python
monthly_mix.plot(figsize=(14, 6))
plt.title("Clean vs Fossil Fuel Share Over Time (Monthly Average)")
plt.xlabel("Month")
plt.ylabel("Average Share (%)")
plt.grid(True)
plt.legend(title="Fuel Group")
plt.tight_layout()
plt.show()
```

## Insight 7: Overall Fuel Share (2024‚Äì2025)

To conclude the analysis, we calculate the total contribution of each fuel type over the entire 2024‚Äì2025 period.

This pie chart provides a snapshot of the UK's energy landscape in aggregate ‚Äî useful for understanding dominant sources and evaluating long-term trends.

- **Wind and gas** are the leading contributors, together accounting for the majority of generation.
- **Nuclear** and **imports** also form a significant share, acting as baseload stabilisers.
- **Solar** and **biomass** remain meaningful but secondary sources.



```python
# Filter data for 2024‚Äì2025 only
filtered_df = clean_df[(clean_df["timestamp_hour"].dt.year >= 2024) & 
                       (clean_df["timestamp_hour"].dt.year <= 2025)]

# Total percentage contribution of each fuel
fuel_totals = filtered_df.groupby("fuel")["perc"].sum()

# Normalize to get share in %
fuel_share = (fuel_totals / fuel_totals.sum()) * 100
fuel_share = fuel_share.sort_values(ascending=False)

# Plot as pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    fuel_share, 
    labels=fuel_share.index, 
    autopct="%1.1f%%", 
    startangle=140,
    wedgeprops={"edgecolor": "black"}
)
plt.title("Overall Fuel Share (2024‚Äì2025)")
plt.tight_layout()
plt.show()
```
