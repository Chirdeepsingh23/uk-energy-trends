```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the latest generation mix CSV
df = pd.read_csv("../data/generation_mix_latest.csv")

# Display the data
df.head()
```

```python
# Bar chart of generation mix
plt.figure(figsize=(10, 6))
plt.bar(df["fuel"], df["perc"])
plt.title("Current UK Electricity Generation Mix")
plt.xlabel("Fuel Type")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
