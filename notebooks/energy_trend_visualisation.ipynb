```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the latest generation mix CSV
df = pd.read_csv("../data/generation_mix_latest.csv")

# Display the data
df.head()
```

```python
# Bar chart of generation mix
plt.figure(figsize=(10, 6))
plt.bar(df["fuel"], df["perc"])
plt.title("Current UK Electricity Generation Mix")
plt.xlabel("Fuel Type")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the historical data
df = pd.read_csv("../data/generation_mix_history.csv")

# Convert timestamp column to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

df.head()


```

```python
# Pivot so each fuel becomes a column
pivot_df = df.pivot(index="timestamp", columns="fuel", values="perc")

# Plot all fuel types over time
pivot_df.plot(figsize=(12, 6))
plt.title("UK Electricity Generation Mix Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Percentage of Total Generation")
plt.grid(True)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
